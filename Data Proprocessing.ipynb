```python
from config import *
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Before running read_csv:
# download csv files from kaggle competition, unzip, and place in csv folder
# may need to change director separators from '/' to '\' if on mac or linux 
train_ratings = pd.read_csv('csv/train_ratings.csv',sep=',')

aveRatingPerMovie = train_ratings.groupby('movieId').mean()
del aveRatingPerMovie['userId']

movies = pd.read_csv('csv/movies.csv',sep=',')
# processes movie stats in movieStats dataframe
# add columns for each genre
# add average score
moviesStats = pd.DataFrame()
moviesWithAveRating = movies.join(aveRatingPerMovie,on='movieId')
# default to rating of 3 if no ratings
moviesWithAveRating = moviesWithAveRating.fillna(3)
moviesStats['Average Rating'] = moviesWithAveRating['rating']

def findYear(title):
    titleRev = title[::-1]
    start=titleRev.find('(')
    end=titleRev.find(')')
    year = titleRev[end+1:start][::-1]
    try:
        return int(year)
    except:
        return 2000 # random guess if rare case no year in title

moviesStats['Year'] = np.vectorize(findYear)(movies['title'])
def getGenres(string,genre):
    return string.find(genre) != -1

possible_genres = [
"Action",
"Adventure",
"Animation",
"Children's",
"Comedy",
"Crime",
"Documentary",
"Drama",
"Fantasy",
"Film-Noir",
"Horror",
"Musical",
"Mystery",
"Romance",
"Sci-Fi",
"Thrillerv",
"War",
"Western",
"(no genres listed)"]
for genre in possible_genres:
    moviesStats[genre] = np.vectorize(getGenres)(movies['genres'], genre)

print("Centralizing movie info into movies stats")
print("Movies Stats dataframe shape: {}".format(moviesStats.shape))
print(moviesStats.describe)

# maybe TODO add tags to movie stats
```
