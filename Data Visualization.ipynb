```python
from config import *
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Before running read_csv:
# download csv files from kaggle competition, unzip, and place in csv folder
# may need to change director separators from '/' to '\' if on mac or linux 
train_ratings = pd.read_csv('csv/train_ratings.csv',sep=',')
```

```python
aveRatingPerMovie = train_ratings.groupby('movieId').mean()
del aveRatingPerMovie['userId']
plt.title('Average Rating per Movie')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.hist(aveRatingPerMovie['rating'],bins=30)
plt.show()
```

```python
plt.hist(train_ratings['rating'])
plt.title('General Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
numUsers = train_ratings.groupby('userId').mean().shape[0]
numRatings = train_ratings.shape[0]
print('Number of Users in Training Set:', numUsers)
print('Number of Ratings in Training Set:', numRatings)
dist = np.unique(train_ratings['userId'],return_counts=True)[1]
plt.title('Number of Ratings per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Frequency')
plt.hist(dist, bins=100)
plt.show()
```
